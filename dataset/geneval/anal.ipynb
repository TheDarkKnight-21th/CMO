```python
import json
from collections import Counter

# 파일 경로 (본인의 파일명으로 변경하세요)
file_path = "train_metadata.jsonl"

def calculate_tag_ratio(file_path):
    tag_counts = Counter()
    total_count = 0
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                try:
                    data = json.loads(line)
                    # 'tag' 키가 있는지 확인 후 카운트
                    if "tag" in data:
                        tag_counts[data["tag"]] += 1
                        total_count += 1
                except json.JSONDecodeError:
                    continue  # 깨진 줄은 건너뜀

        # 결과 출력
        print(f"--- 총 데이터 개수: {total_count}개 ---")
        for tag, count in tag_counts.most_common():
            ratio = (count / total_count) * 100
            print(f"[{tag}]: {count}개 ({ratio:.2f}%)")

    except FileNotFoundError:
        print("파일을 찾을 수 없습니다.")

# 실행
calculate_tag_ratio(file_path)
```

```python
import json
from collections import defaultdict

# 파일 이름 설정
input_file = "train_metadata.jsonl"

# 어떤 키가 나올지 모르므로, 딕셔너리 안에 세트(set)를 자동으로 생성하도록 설정
unique_values_per_key = defaultdict(set)

print(f"[{input_file}] 파일의 모든 키와 고유값을 탐색합니다...\n")

with open(input_file, 'r', encoding='utf-8') as f:
    for idx, line in enumerate(f):
        line = line.strip()
        if not line:
            continue
            
        try:
            data = json.loads(line)
            
            # include 배열 안에 있는 객체들을 하나씩 검사
            for item in data.get("include", []):
                
                # 객체 안의 모든 키(key)와 값(value)을 동적으로 순회!
                for key, value in item.items():
                    
                    # 리스트(예: position의 ["below", 0])나 딕셔너리는 set에 넣을 수 없으므로 튜플이나 문자열로 변환
                    if isinstance(value, list):
                        value = tuple(value) # 리스트 -> 튜플로 변환해서 저장
                    elif isinstance(value, dict):
                        value = str(value)   # 딕셔너리 -> 문자열로 변환해서 저장
                        
                    # 해당 키의 set에 값을 추가 (중복은 알아서 제거됨)
                    unique_values_per_key[key].add(value)
                    
        except json.JSONDecodeError:
            print(f"⚠️ 경고: {idx + 1}번째 줄 문법 에러로 건너뜁니다.")
            continue

# 수집된 모든 키와 고유값 결과를 깔끔하게 출력
print("================ 탐색 결과 ================\n")

# 발견된 키들을 알파벳 순서대로 정렬하여 출력
for key in sorted(unique_values_per_key.keys()):
    values = unique_values_per_key[key]
    
    # 값들도 정렬 (문자열, 숫자, 튜플 등이 섞여 있을 수 있으므로 모두 문자열 기준으로 정렬)
    sorted_values = sorted(list(values), key=str)
    
    print(f"🔑 [{key}] (총 {len(sorted_values)}개의 고유값)")
    for v in sorted_values:
        print(f"  - {v}")
    print("-" * 40)
```
