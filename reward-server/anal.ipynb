```python
import torch

def print_param_count(model_name, model):
    if model is None:
        print(f"{model_name}: 모델을 불러오지 못했습니다.")
        return
    
    # 총 파라미터 수 계산
    total_params = sum(p.numel() for p in model.parameters())
    
    # Billion (10억) 단위로 변환
    params_in_b = total_params / 1_000_000_000
    
    # 쉼표가 포함된 실제 개수와 B 단위 출력
    print(f"{model_name}: {total_params:,} ({params_in_b:.2f}B)")


def check_model_parameters():
    print("모델 파라미터 계산을 시작합니다...\n" + "="*50)

    # 1. SAM3
    try:
        from transformers import Sam3Model
        print("[1/3] SAM3 로드 중...")
        sam_model = Sam3Model.from_pretrained("facebook/sam3")
        print_param_count("SAM3", sam_model)
        del sam_model # 메모리 관리를 위해 삭제
    except Exception as e:
        print(f"SAM3 로드 실패: {e}")

    # 2. Standard OpenCLIP (ViT-H-14)
    try:
        import open_clip
        print("\n[2/3] OpenCLIP (ViT-H-14) 로드 중...")
        clip_model, _, _ = open_clip.create_model_and_transforms('ViT-H-14', pretrained='laion2b_s32b_b79k')
        print_param_count("OpenCLIP (ViT-H-14)", clip_model)
        del clip_model
    except Exception as e:
        print(f"OpenCLIP 로드 실패: {e}")

    # 3. Depth Anything V3 (da3-small)
    try:
        from depth_anything_3.api import DepthAnything3
        print("\n[3/3] Depth Anything V3 로드 중...")
        depth_model = DepthAnything3.from_pretrained("depth-anything/da3-small")
        # DepthAnything3 객체가 래퍼(wrapper) 클래스라면 내부 모델에 접근해야 할 수도 있습니다.
        # getattr를 사용하여 내부 nn.Module에 안전하게 접근합니다.
        actual_model = getattr(depth_model, 'model', depth_model) 
        print_param_count("Depth Anything V3 (Small)", actual_model)
        del depth_model
    except Exception as e:
        print(f"Depth Anything V3 로드 실패: {e}")

    print("="*50 + "\n완료되었습니다.")

if __name__ == "__main__":
    check_model_parameters()
```
